**WORKING ON THE JOINED COLUMN**

```python
from datetime import datetime
fifa['Joined']
```

```python
def month_to_number(x):
            if x== 'Jan':
                return 1
            elif x == 'Feb':
                return 2
            elif x == 'Mar': 
                return 3
            elif x == 'Apr':
                return 4
            elif x == 'May':
                return 5
            elif x == 'Jun': 
                return 6
            elif x == 'Jul': 
                return 7
            elif x == 'Aug': 
                return 8
            elif x == 'Sep': 
                return 9
            elif x == 'Oct': 
                return 10
            elif x == 'Nov':  
                return 11
            elif x == 'Dec':  
                return 12
```

```python
date = []
for x in range(len(fifa['Joined'])):
    d = fifa['Joined'][x]
    c = d.split(" ")
    month = str(month_to_number(c[0]))
    day = str(c[1].replace(',',''))
    year = str(c[2])
    if len(day) == 1:
        day = ('0'+str(day))
    date_long = (str(month)+'/'+str(day)+'/'+str(year))
    date_con = pd.to_datetime(date_long)
    date.append(date_con)
fifa['Joined'] = date
fifa['Joined']
```

```python
fifa.head()
```

```python
#TO ENSURE CONSISTENCY IN NAMING, OVA IS REMANED
fifa = fifa.rename(columns = {'↓OVA': 'OVA'})
```

```python
#Cleaning up the club dataset
fifa['Club']= fifa['Club'].str.replace('\n\n\n\n','')
fifa.tail(6)
```

**The next Column to clean is the contract**

```python
fifa.Contract.unique()
```

**Observations: Some players are on contract while some are on loan, in other to make the column type uniform, we give players with active contract as active while the ones on loan should be categorized as on loan and others free agents**

```python
# define a function to change the values with 'on loan' as an "On loan" category
# define the players with active loans as Active and they can be located easily cause they contain '~'
def contract(a):
    if "On Loan" in a:
        a = 'ON LOAN'
        return a
    elif '~' in a:
        a = 'ACTIVE'
        return a
    elif 'Free' in a:
        a = "FREE"
        return a
```

```python
fifa.Contract = fifa.Contract.apply(contract)
fifa.Contract = fifa.Contract.astype('category')
# Rename the column to Contract status
fifa = fifa.rename(columns={'Contract':'ContractStatus'})
fifa.head()
```

**CLEANING UP POSITIONS COLUMN**

```python
#Ordering position column leaving them in groups
temp_position = []
for x in range(len(fifa['Positions'])):
    y = sorted(fifa['Positions'][x].split(" "))
    yx = ' '.join(y)
    temp_position.append(yx)

fifa['Positions'] = temp_position
fifa['Positions'] = fifa['Positions'].astype(object)
fifa['Positions'].unique()
```

**WORKING ON THE W/F,SM and IR COLUMNS**

**The W/F,SM and IR columns have ★ in them and it should be removed**

```python
#Defining afunction to help remove star
def remove_star(x):
    x = x.replace('★', '')
    return x
```

```python
fifa.IR = fifa.IR.apply(remove_star)
fifa.SM = fifa.SM.apply(remove_star)
fifa['W/F'] = fifa['W/F'].apply(remove_star)
fifa.head()
```

The next step is to clean the Height and Weight columns as they are both cast as an object type

```python
fifa.Height.unique()
```

Notice from the data, we have "cm" and also \'"

```python
def convert_height(x):
    if x[-1] == '"':
        x = x.replace("\"","")
        inch = int(x[2:]) * 2.54
        foot = int(x[0]) * 30.48
        return round(foot+inch)
    elif x[-1] == "m":
        return int(x[:-2])
```

```python
fifa.Height = fifa.Height.apply(convert_height)
fifa.Height.unique()
```

```python
fifa = fifa.rename(columns={'Height':'Height_cm'})
```

```python
fifa.Weight.unique()
```

```python
def weight_kg(x):
    if 'lbs' in x:
        x = x.replace('lbs','')
        x = float(x) * 0.45359237
        x = int(x)
        return x
    else:
        return int(x[:-2])
```

```python
fifa.Weight = fifa.Weight.apply(weight_kg)
fifa = fifa.rename(columns={'Weight':'Weight_cm'})
```

**The next task is to clean up the release column and remame the column**

```python
fifa["Release Clause"].unique()
```

```python
def release(x):
    if 'M' in x:
        x = x.replace('M', '')
        x = float(x[1:]) * 1000000
        return round(x)
    elif "K" in x:
        x = x.replace('K', '')
        x = float(x[1:]) * 1000
        return round(x)
    else:
        return x
```

```python
fifa["Release Clause"] = fifa["Release Clause"].apply(release)
fifa["Release Clause"].unique()
```

```python
fifa = fifa.rename(columns={'Release Clause':'Release Clause_€'})
```

**VALUE**

```python
def value(x):
    if 'M' in x:
        x = x.replace('M', '')
        x = float(x[1:]) * 1000000
        return round(x)
    elif "K" in x:
        x = x.replace('K', '')
        x = float(x[1:]) * 1000
        return round(x)
    else:
        return int(x[1:])
```

```python
fifa.Value = fifa.Value.apply(value)
```

**HITS**

```python
fifa.Hits.unique()
```

**Observations: You will notice that some data has K in them, we will need to clean it**

```python
# define a function to replace K with an empty string and turn the value to a float so it can be added to 1000
def hitchange(h):
    if "K" in h:
        h = h.replace('K', "")
        h = float(h)* 1000
        return round(h)
    else:
        return h
```

```python
fifa.Hits = fifa.Hits.apply(hitchange)
fifa.Hits.unique()
```

```python
#converting the datatype from object to integer
fifa['Hits'] = fifa['Hits'].astype('int64')
```
